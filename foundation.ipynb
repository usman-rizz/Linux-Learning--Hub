<h2 style="color:#2E7D32;  margin-top:0;">
     <b> What is Linux basically? </b>
  </h2>


Linux is an open source `Unix-like operating system kernel` that manages system resources (CPU, memory, and hardware devices) and allows applications to interact with hardware. 

It is widely used in servers, cloud infrastructure, and big data platforms because of its stability, security, and flexibility.

In cloud and big data environments like Apache Hadoop, Apache Spark, and Apache Airflow, are designed to run efficiently on Linux environments. 

Linux is the standard OS.

Different operating systems such as `Ubuntu` or `Debian` are built on top of the Linux kernel.


<h2 style="color:#2E7D32;  margin-top:0;">
     <b> What is UNIX ?  </b>
  </h2>



Unix is a very `old operating system` developed in 1969 at Bell Labs.

It became the foundation of many modern operating systems. 
Unix introduced many concepts still used today such as:

`hierarchical file system, multi-user system, command-line interfaces, shell commands...`

and if we talk about `Unix based OS` it means an operating system that follows the design principles of Unix.

Examples:

Linux
macOS
FreeBSD

all these 3 follows the architecture of Unix..


<h2 style="color:#2E7D32;  margin-top:0;">
     <b> What is a Kernel? </b>
  </h2>




A `kernel` is the core component of an operating system. 

It manages communication between `software applications and hardware.`
Whenever a program needs CPU, memory, or disk access, the kernel handles that request and allocates resources.


**Example**

when we run a program like Python, the application requests resources such as CPU and memory.
The kernel receives that request and allocates the required resources so the program can execute properly.


<h2 style="color:#2E7D32;  margin-top:0;">
     <b> Linux Architecture  </b>
  </h2>




Understanding `Linux architecture` is important because it helps to know EXACTLY what happens when I run a command or a pipeline.

When something breaks in production, I don't just guess, I know which layer the problem is in. 
Is it the application code? Is it the kernel running out of memory? Is it the hardware disk that's full? 

Without knowing the architecture, you're just randomly trying fixes. With it, you debug in minutes instead of hours.

Understand WHERE Errors Come From..

When your `Spark job crashes`, there are 4 possible places the error lives:

 Application Layer  →  Bug in your Python/Spark code
 Shell Layer        →  Wrong command, bad script syntax
 Kernel Layer       →  Out of memory, too many processes
 Hardware Layer     →  Disk full, disk I/O too slow

A person who doesn't know architecture sees one error message and panics.
You will look at it and say:

"This is an OOM, `Out of Memory error`. That's a Kernel-level issue. The Kernel tried to allocate RAM for my Spark job but couldn't. 
I need to increase executor memory or kill other processes."

so knowing the architecture is must...

<h2 style="color:#2E7D32;  margin-top:0;">
     <b> Linux Distro  </b>
  </h2>




A `Linux Distribution` or `distro` is a complete operating system (Linux Kernel + Shell + GNU Tools + Package Manager) built ON TOP of the Linux Kernel bundled together for a specific purpose and audience. 

All distros share the same kernel but differ in tools, stability, and use case. 


Linux Kernel = The ENGINE (same in every distro)\
Distribution  = ENGINE + Body + Features + Tools


Ubuntu          = Linux Kernel, beginner friendly, desktop/server\
CentOS          = Linux Kernel, enterprise stable, production servers\
Kali Linux      = Linux Kernel, security tools, ethical hacking

***Same engine. Different purpose. Different tools.***

<h2 style="color:#37474F; text-align:center; margin-top:0;">
    Most Important Linux Distributions (Distros)
  </h2>


### **Ubuntu**

Ubuntu is one of the most popular `Linux distributions`. It is beginner friendly, has `long-term support (LTS) versions`, and is widely used in *servers and cloud platforms*.\
If you are a data engineer or DevOps professional, you will most likely use Ubuntu because it works out-of-the-box with Python, Spark, Docker, and cloud platforms like AWS or GCP.


### **Debian**

Debian is the foundation of many other distros including Ubuntu.\
It is extremely stable but doesn’t get the latest updates as fast. Ideal for production servers where you want reliability.


### **CentOS / RHEL**

CentOS is like a free version of `Red Hat Enterprise Linux`. It’s very stable, and used for enterprise production servers.\
RHEL is paid and comes with official support.

**Usage:**
Banks, government servers, corporate environments.

### **RHEL (Red Hat Enterprise Linux)**

Paid enterprise version with professional support.\
Critical for enterprise systems where downtime is not acceptable.


### **Kali Linux**

Special purpose distro for `ethical hacking` and `security auditing`.\
Comes with hundreds of security tools pre-installed. Not for general use.


### **Linux Mint**

Ubuntu-based, very user-friendly, great GUI for beginners.\
If someone is switching from Windows to Linux, Mint is easiest.
